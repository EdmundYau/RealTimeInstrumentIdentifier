```python
import numpy as np

# Load the .npy file
data = np.load("labels_reduced_test/Track01877.npy")


# Instrument mapping based on the provided image
instrument_mapping = {
    0: "Piano",
    1: "Guitar",
    2: "Bass",
    3: "Strings",
    4: "Brass",
    5: "Reed",
    6: "Pipe",
    7: "Drums"
}

# Process and print formatted output
num_seconds = data.shape[1]  # Number of time steps

for sec in range(num_seconds):
    active_instruments = [instrument_mapping[i] for i in range(data.shape[0]) if data[i, sec] == 1]
    instruments_str = " ".join(active_instruments)
    print(f"Detected Instruments at second {sec}: {instruments_str}")
```
